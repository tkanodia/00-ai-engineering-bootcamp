```python
from langsmith import Client
```

### Attach Feedback to a specific trace_id in Langsmith
#### Create a trace and use ID to perform this test

trace id - f05484a3-d7e3-4aee-b373-eeeafd89cdd1

```python
client = Client()
```

```python
traceId = "f05484a3-d7e3-4aee-b373-eeeafd89cdd1"
```

```python
client.create_feedback(
    run_id=traceId,
    key="feedback",
    score=0,
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=traceId,
    key="feedback_1",
    score=1,
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=traceId,
    key="comment",
    value="This is a long form feedback you should use to improve the system!",
    feedback_source_type="api"
)
```
